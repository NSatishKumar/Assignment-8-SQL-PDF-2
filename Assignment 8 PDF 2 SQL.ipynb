## 1. Create an sqlalchemy engine using a sample from the data set

```python
# Importing the packages
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Reading the csv file and storing in a dataframe
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data',sep=',',header=None,index_col=False)
df
```

```python
# Renaming the column names
colVal=["age","workclass","fnlwgt","education","education_num","matrial_status","occupation","relationship","race","sex","capital_gain","capital_loss","hours_per_week","native_country","income"]
df.columns =colVal
```

```python
df.head()
```

```python
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import Column, Integer, String
Base = declarative_base()

class adult(Base):                                            # Declaring the class adult with all columns
        __tablename__ = 'adults'

        id=Column(Integer, primary_key=True)
        age=Column(Integer)
        workclass=Column(String(250))
        fnlwgt=Column(Integer)
        education = Column(String(250))
        education_num = Column(Integer)
        matrial_status =Column(String(250))
        occupation=Column(String(250))
        relationship=Column(String(250))
        race = Column(String(250))
        sex = Column(String(250))
        capital_gain = Column(Integer)
        capital_loss=Column(Integer)
        hours_per_week=Column(Integer)
        native_country = Column(String(250))
        income = Column(String(250))
```

```python
engine = create_engine('sqlite:///Data.db' , echo=True)      # Declaring the engine
Base.metadata.create_all(engine)
```

```python
df.to_sql('adults', engine,  index=True, index_label='id', if_exists='append')  # storing into SQL 
```

```python
engine.execute('select count(*) from adults').fetchall()       # Fetching all records
```

```python
from sqlalchemy.orm import sessionmaker
Session = sessionmaker()                               # Defining the session
Session.configure(bind=engine) 
session = Session()
# Displaying the Id and age of all records
for row in session.query(adult).filter_by(workclass=' State-gov').all():
    print(row.id,row.age)
```
